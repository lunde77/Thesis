```python
import os

# Specify the folder path
folder_path = base_path

# Get a list of all files in the folder
file_names = os.listdir(folder_path)
```

```python
import os
import pandas as pd

# Create a dictionary to store the loaded dataframes
dataframes = {}

# Create a list to store the dataframe names
dataframe_names = []

# Load the CSV files and store them in the dictionary
for file_name in file_names:
    file_path = os.path.join(folder_path, file_name)
    dataframe_name = os.path.splitext(file_name)[0][3:]  # Remove the ".csv" extension and the first 3 characters, so only the CD numer is there
    dataframes[dataframe_name] = pd.read_csv(file_path)
    dataframe_names.append(dataframe_name)

```

# get intial data summarys 

Calculate some summary statistics for the time the CD are connected

```python
# Initialize lists to store the statistics for each dataframe
averages = []

# Iterate through the dataframes and calculate statistics
for dataframe_name, dataframe in dataframes.items():
    connected_column = dataframe["connected (1 or 0)"]
    
    # Calculate the average (mean)
    average = connected_column.mean()
    averages.append(average)
    
# Calculate quartiles (Q1 and Q3)
q1 = np.percentile(averages, 25)
q3 = np.percentile(averages, 75)
median = np.percentile(averages, 50)

# Calculate the maximum and minimum values
max_value = np.max(averages)
min_value = np.min(averages)

# Calculate the average (mean) of the averages (meta-average?)
average_of_averages = np.mean(averages)

# Print the results
print("Median of Averages:", median)
print("Q1 of Averages:", q1)
print("Q3 of Averages:", q3)
print("Max of Averages:", max_value)
print("Min of Averages:", min_value)
print("Average of Averages:", average_of_averages)
```
